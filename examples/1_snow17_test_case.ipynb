```python
filename = filename1
fullpath = os.path.join(filepath, filename)

# df = pd.read_csv(fullpath, delim_whitespace=True, header=0) # use delimiter = '\t' or ',' if comma-separated 
df = pd.read_csv(fullpath, sep=r'\s+', header=0) # using delim_whitespaced worked for the header, but it made a warning so the following is a work around

df
```

```python
print(df.columns)
```

```python
# change the name 'mo' to 'month, 'dy' to 'day', 'hr' to 'hour'
df.rename(columns={'mo':'month', 'dy':'day', 'hr':'hour'}, inplace=True)
df.head()
```

```python
# now, create a datetime object from the year, month, day, hour, minute columns
df['datetime'] = pd.to_datetime(df[['year', 'month', 'day', 'hour']])
# index by datetime 
df.set_index('datetime', inplace=True)
df.head()
```

```python
# now, plot the snowh with datetime 
plt.figure(figsize=(12,6))
plt.plot(df['sneqv'])
plt.xlabel('Date')
plt.ylabel('swe (mm)')
plt.title('Snow Depth at HHWM8IL')
plt.show()
```
